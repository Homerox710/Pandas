## Group by

```
df = sns.load_dataset('diamonds')
df
```

```
df.groupby('cut').mean()
```

```
df.groupby('cut').median()
```

```
df.groupby('cut')['carat'].count()
```

```
df.groupby('cut')['carat'].max()
```

**Vamos a recorrer los valores, agrupándolos por el corte luego vamos a calcular el promedio de cada uno**

```
for key_group, group in df.groupby('cut'):
  grouped_price = group['price'].mean()
  print(f"Cut: {key_group}, Price {grouped_price} \n")
```

**Agrupar con varios parámetros**

```
df.groupby(['cut','color'])['price'].mean().to_frame().head(10)
```

## Aggregate

**Con aggregate podemos definir una lista con todas las funciones a utilizar**

El group by representa los índices y los elementos de aggregate, las columnas

```
df.groupby(['cut','color'])['price'].aggregate(['min',np.mean,max]).head(10)
```

**Podemos usar las funciones por defecto o crear las nuestras**

```
def mean_kilo(x):
  return np.mean(x)/1000
```

**Las funciones se aplican a todos los grupos**

```
df.groupby(['cut','color'])['price'].aggregate(['min',np.mean,mean_kilo]).head(10)
```

**Podemos crear un diccionario de instrucciones y colocarlo en aggregate**

```
dict_agg = {'carat':[min,max],
            'price':[np.mean, mean_kilo]
            }
dict_agg
```

```
df.groupby(['cut','color']).aggregate(dict_agg).head(10)
```

## Filtros

**Función para hacer filtros**

```
# Reliza el calculo del promedio y retorna sólo los mayores a 4
def f_filter(x):
  return mean_kilo(x['price'])>4
```

**Retorna sólo los que su precio promedio dividido entre 1000 es mayor a 4**

```
df.groupby('cut').filter(f_filter).head(10)
```

Si vemos sólo 2 tipos de corte cumplen la condición, ya que son los más costosos

```
df.groupby('cut').filter(f_filter)['cut'].unique()
```

## Apply

Función para transformae euros a dólares

```
def mean_eur2usd(x):
    y = np.mean(x)*1.12
    return y
```

```
mean_eur2usd([100,102])
```

```
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(np.std)
```

Apply acepta funciones predeterminadas, las que creamos y también lambdas

```
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

## Aggegate

```
df.groupby(['sex','time'])[['total_bill','prct_tip']].aggregate([np.mean,np.max])
```

Aggregate recibe una lista con las funciones que queremos aplicar

```
df.groupby(['sex','time'])[['total_bill','prct_tip']].agg([np.mean,np.max])
```

También acepta un diccionario cómo parámetro

```
dict_agg = {'tip':[min, max],'total_bill':[np.mean, mean_eur2usd]}
dict_agg
```

```
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
```

## Filter

```
def f_filter(x):
    return mean_eur2usd(x['total_bill'].mean()) > 20
```

```
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
```

```
df_filtered.groupby(['sex','time']).count()
```

## Extraer valor con variables categóricas

```
df
```

Creamos una columna con unos

```
df['ones'] = 1
```

```
df
```

Agrupo las categorías ayudandonos aplicándole una suma a la columna ones

```
df_g = df.groupby(['sex','smoker'])[['ones']].sum()
df_g
```

Ahora aplicamos esta función sobre ones para calcular los porcentajes

```
df_g.groupby(level=0).apply(

lambda x:
x / x.sum() * 100

)
```

Otra forma de extraer valor de los datos (continuos en este caso) es categorizándola en grupos dentro de su rango

```
# Cut permite hacer esta categorización, recibe la variable y una lista de categorías
df['bin_total'] = pd.cut(df['total_bill'],bins = [3,18,35,60])
df
```

```
pd.cut(df['total_bill'],bins = [3,18,35,60]).value_counts()
```

Distribución de los valores en los tiempos de comida

```
df.groupby(['time','bin_total'])[['ones']].count()
```

Distribución en porcentajes

```
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply(

lambda x:
x / x.sum() * 100

)
```

# Pivot

```
df.groupby(['sex','time'])[['total_bill']].mean()
```

Reseteamos los índices para manipularlos

```
df_gp = df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp
```

## Tabla dinámica

Cambiamos los índices a columnas

```
df_gp.pivot_table(values='total_bill', index='sex', columns='time')
```

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

Pivot sobre todo el df

```
df.pivot_table(values='total_bill', index='sex', columns='time')
```

Pivot ejecuta mean por default, podemos usar aggfunc para cambiar esto

```
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=np.median)
```

```
df_pivot = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median,np.std])
df_pivot
```

Reseteamos los índices para ver mejor la función que hace pivot sobre el df

```
df_pivot.unstack().reset_index()
```
